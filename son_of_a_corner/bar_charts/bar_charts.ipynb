## Percentage of fouls conceded and drawn by teams

```python
fouls_per_team_venue = (
    df[df['variable'] == 'fouls_for']
    .groupby(['team_name', 'venue'])
    ['value'].sum()
    .reset_index()
)

sort_order = fouls_per_team['team_name'].to_list()
fouls_per_team_venue = fouls_per_team_venue.set_index('team_name')
fouls_per_team_venue = fouls_per_team_venue.loc[sort_order]
fouls_per_team_venue = fouls_per_team_venue.reset_index()

height_h = (
    fouls_per_team_venue[fouls_per_team_venue["venue"] == "H"]["value"]
    .reset_index(drop = True)
)

height_a = (
    fouls_per_team_venue[fouls_per_team_venue["venue"] == "A"]["value"]
    .reset_index(drop = True)
)

X = np.arange(len(height_h))
```

```python
fig = plt.figure(figsize = (6, 2.5), dpi = 200, facecolor = "#EFE9E6")
ax = plt.subplot(111, facecolor = "#EFE9E6")

ax.spines[['top', 'right']].set_visible(False)

ax.grid(color = "grey", ls = ":", lw = 0.5)
ax.set_ylabel("Fouls conceded")

ax.bar(
    X,
    height_h,
    ec = "black",
    lw = .75,
    width = .75,
    color = "#005a9b",
    zorder = 3,
    label = "Home"
)

ax.bar(
    X,
    height_a,
    bottom = height_h,
    ec = "black",
    lw = .75,
    width = .75,
    color = "#f4a261",
    zorder = 3,
    label = "Away"
)

ax.legend(
    ncol = 2, 
    loc = "upper center", 
    bbox_to_anchor = (0.45, 1.2),
    frameon = False,
    fontsize = 10
)

index = 0
for y_home, y_away in zip(height_h, height_a):
    home_text = ax.annotate(
        xy = (index, y_home/2),
        text = f'{y_home/(y_home + y_away):.0%}',
        size = 7,
        ha = "center",
        va = "center",
        weight = "bold",
        color = "white",
        rotation = 90
    )

    away_text = ax.annotate(
        xy = (index, y_home + y_away/2),
        text = f'{y_away/(y_home + y_away):.0%}',
        size = 7,
        ha = "center",
        va = "center",
        weight = "bold",
        color = "white",
        rotation = 90
    )

    home_text.set_path_effects(
        [path_effects.Stroke(linewidth=1.75, foreground="black"), path_effects.Normal()]
    )

    away_text.set_path_effects(
        [path_effects.Stroke(linewidth=1.75, foreground="black"), path_effects.Normal()]
    )
    
    index += 1


ax.xaxis.set_ticks(
    labels = sort_order,
    rotation = 90,
    ticks = X
)

plt.show()
```

## Fouls conceded and drawn per team

```python
fouls_per_team_c = (
    df[df['variable'] == 'fouls_for']
    .groupby(['team_name'])
    ['value'].sum()
    .reset_index()
)

fouls_per_team_d = (
    df[df['variable'] == 'fouls_ag']
    .groupby(['team_name'])
    ['value'].sum()
    .reset_index()
)

fouls_per_team_c = fouls_per_team_c.sort_values(['value'], ascending=[False])
sort_order = fouls_per_team_c['team_name'].to_list()
fouls_per_team_d = fouls_per_team_d.set_index('team_name')
fouls_per_team_d = fouls_per_team_d.loc[sort_order]
fouls_per_team_d = fouls_per_team_d.reset_index()

height_c = (
    fouls_per_team_c["value"].reset_index(drop = True)
)

height_d = (
    fouls_per_team_d["value"].reset_index(drop = True)
)

X = np.arange(len(height_c))
```

```python
fig = plt.figure(figsize = (6.5, 2.5), dpi = 200, facecolor = "#EFE9E6")
ax = plt.subplot(111, facecolor = "#EFE9E6")

ax.spines[['top', 'right']].set_visible(False)

ax.grid(color = "grey", ls = ":", lw = 0.5)

width = 0.35

ax.bar(
    X,
    height_c,
    width = width,
    color = "#005a9b",
    ec = "black",
    lw = .75,
    zorder = 3,
    label = "Fouls conceded"
)

ax.bar(
    X + width,
    height_d,
    width = width,
    color = "#f4a261",
    ec = "black",
    lw = .75,
    zorder = 3,
    label = "Fouls drawn"
)

ax.xaxis.set_ticks(
    ticks = X + width/2,
    labels = sort_order,
    rotation = 90
)

ax.legend(
    ncol = 2, 
    loc = "upper center", 
    bbox_to_anchor = (0.45, 1.2),
    frameon = False,
    fontsize = 9
)

plt.show()
```

## Foul conceded to drawn ratio per team

```python
ratio_df = pd.DataFrame(
    {
        "team_name":sort_order,
        "fouls_drawn":height_d,
        "fouls_conceded":height_c
    }
)

ratio_df["ratio"] = ratio_df["fouls_conceded"]/ratio_df["fouls_drawn"]
ratio_df = ratio_df.sort_values(["ratio"], ascending=[False])

height = ratio_df["ratio"].reset_index(drop=True)
X = np.arange(len(height))
```

```python
fig = plt.figure(figsize = (6, 2.5), dpi = 200, facecolor = "#EFE9E6")
ax = plt.subplot(111, facecolor = "#EFE9E6")

ax.spines[['top', 'right']].set_visible(False)
ax.grid(color = "grey", ls=":", lw = 0.5)

width = 0.65

ax.bar(
    X,
    height,
    width = width,
    ec = "black",
    lw = .75,
    color = "#005a9b",
    zorder = 3,
    label = "Fouls conceded"
)

for index, y in enumerate(height):
    ax.annotate(
        xy = (index, y),
        text = f'{y:.2f}',
        xytext = (0, 7),
        textcoords = "offset points",
        size = 7,
        ha = "center",
        va = "center",
        weight = "bold",
        color = "#005a9b"
    )

ax.xaxis.set_ticks(
    ticks = X,
    labels = ratio_df["team_name"],
    rotation = 90
)

plt.show()
```
