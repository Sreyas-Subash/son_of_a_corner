## Data Engineering

```python
home_df = df.copy()
home_df = home_df.melt(id_vars=['date', 'home_team_name', 'away_team_name'])
home_df['venue'] = 'H'
home_df.rename(columns={'home_team_name':'team', 'away_team_name':'opponent'}, inplace=True)
home_df.replace({'variable':{'home_team_xG':'xG_for', 'away_team_xG':'xG_ag'}}, inplace=True)
```

```python
home_df
```

```python
away_df = df.copy()
away_df = away_df.melt(id_vars=['date', 'away_team_name', 'home_team_name'])
away_df['venue'] = 'A'
away_df.rename(columns={'away_team_name':'team', 'home_team_name':'opponent'}, inplace=True)
away_df.replace({'variable':{'away_team_xG':'xG_for', 'home_team_xG':'xG_ag'}}, inplace=True)
```

```python
away_df
```

```python
df = pd.concat([home_df, away_df]).reset_index(drop=True)
```

```python
df.info()
```

```python
df = df[df['team'] == 'Brighton & Hove Albion']
df = df.sort_values(by = 'date').reset_index(drop=True)
```

```python
df
```

```python
Y_for = df[df['variable'] == 'xG_for'].reset_index(drop=True)
Y_ag = df[df['variable'] == 'xG_ag'].reset_index(drop=True)
X = pd.Series(range(len(Y_for)))
```

```python
Y_for = Y_for['value'].rolling(window = 10, min_periods = 0).mean()
Y_ag = Y_ag['value'].rolling(window = 10, min_periods = 0).mean()
```

## Viz

```python
fig = plt.figure(figsize=(4,2.5), dpi = 200)
ax = plt.subplot(111)

ax.plot(X, Y_for, label = "xG created")
ax.plot(X, Y_ag, label = "xG conceded")

ax.legend()
```

```python
fig = plt.figure(figsize=(4,2.5), dpi = 200)
ax = plt.subplot(111)

ax.spines[['top', 'right']].set_visible(False)
ax.spines[['bottom', 'left']].set_color('grey')

ax.tick_params(
    color = "grey",
    length = 0,
    labelsize = 4,
    labelcolor = "grey",
    which = "major"
)

ax.xaxis.set_major_locator(ticker.MultipleLocator(19))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.set_ylim(0, 2)

ax.grid(
    color = "grey",
    lw = 0.5,
    ls = ":",
    zorder = -1,
    alpha = 0.6
)

ax.plot(X, Y_for, label = "xG created")
ax.plot(X, Y_ag, label = "xG conceded")

ax.legend(fontsize = 6)
```

```python
fig = plt.figure(figsize=(4,2.5), dpi = 200)
ax = plt.subplot(111)

ax.spines[['top', 'right']].set_visible(False)
ax.spines[['bottom', 'left']].set_color('grey')

ax.tick_params(
    color = "grey",
    length = 0,
    labelsize = 4,
    labelcolor = "grey",
    which = "major"
)

ax.xaxis.set_major_locator(ticker.MultipleLocator(19))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.set_ylim(0, 2)

ax.grid(
    color = "grey",
    lw = 0.5,
    ls = ":",
    zorder = -1,
    alpha = 0.6
)

ax.plot(X, Y_for, label = "xG created", zorder = 3)
ax.plot(X, Y_ag, label = "xG conceded", zorder = 3)

ax.fill_between(
    X,
    Y_for,
    Y_ag,
    where = Y_for > Y_ag,
    interpolate = True,
    alpha = 0.75,
    zorder = 2
)

ax.fill_between(
    X,
    Y_for,
    Y_ag,
    where = Y_ag > Y_for,
    interpolate = True,
    alpha = 0.75,
    zorder = 2
)


ax.legend(fontsize = 6)
```

```python
fig = plt.figure(figsize=(4,2.5), dpi = 200)
fig.set_facecolor("#EFE9E6")
ax = plt.subplot(111, facecolor = "#EFE9E6")

ax.spines[['top', 'right']].set_visible(False)
ax.spines[['bottom', 'left']].set_color('grey')

ax.grid(
    color = "grey",
    lw = 0.5,
    ls = ":",
    zorder = -1,
    alpha = 0.6
)

ax.tick_params(
    color = "grey",
    length = 0,
    labelsize = 5,
    labelcolor = "grey",
    which = "major"
)

ax.xaxis.set_major_locator(ticker.MultipleLocator(19))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.set_ylim(0, 2)


line1 = ax.plot(X, Y_for, label = "xG created", color = "#0057B8", zorder = 4)
line2 = ax.plot(X, Y_ag, label = "xG conceded", zorder = 4, color = "#989898")

ax.fill_between(
    X,
    Y_for,
    Y_ag,
    where = Y_for > Y_ag,
    interpolate = True,
    alpha = 0.75,
    zorder = 3,
    color = line1[0].get_color()
)

ax.fill_between(
    X,
    Y_for,
    Y_ag,
    where = Y_ag > Y_for,
    interpolate = True,
    alpha = 0.75,
    zorder = 3,
    color = line2[0].get_color()
)

ax.plot(
    [38, 38],
    [ax.get_ylim()[0], ax.get_ylim()[1]],
    ls = ":",
    lw = 1,
    color = "grey",
    zorder = 2
)

ax.annotate(
    xy = (38, 0.55),
    xytext = (46, 0.75),
    text = "Change in EPL season",
    weight = "light",
    size = 5.5,
    color = "grey",
    arrowprops = dict(
        arrowstyle = "->",
        color = "grey",
        linewidth = 0.75,
        connectionstyle="angle3, angleA=50, angleB=-40",
        # shrinkB = 5,
        shrinkA = 0
    )
)

fig_text(
    annotationbbox_kw = {"xycoords":"figure fraction"},
    x = 0.12,
    y = 1.1,
    color = "black",
    weight = "bold",
    s = "Brighton & Hove Albion"
)

fig_text(
    annotationbbox_kw = {"xycoords":"figure fraction"},
    x = 0.12,
    y = 1.02,
    color = "black",
    size = 6,
    s = "Expected goals <created> and <conceded> | 10-match rolling average\nEPL seasons 20/21 and 21/22",
    highlight_textprops = [
        {"color":line1[0].get_color(), "weight":"bold"},
        {"color":line2[0].get_color(), "weight":"bold"}
    ]
)

url = "https://images.fotmob.com/image_resources/logo/teamlogo/10204.png"
logo_ax = fig.add_axes([0.82, 0.97, 0.13, 0.13], zorder=1)
club_icon = Image.open(urllib.request.urlopen(url))
logo_ax.imshow(club_icon)
logo_ax.axis('Off')

plt.show()
```
